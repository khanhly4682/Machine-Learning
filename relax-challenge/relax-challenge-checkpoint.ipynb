### Users

Let check the number of visits for each user in one week.If an user has less than 3 visits in a week, we remove them.

```python
# Change time_stamp to datetime
engage.time_stamp = pd.to_datetime(engage.time_stamp)
```

```python
# Adopt users that have total visits less than 3 times.
engage_group = engage.groupby('user_id').resample('1W').sum()
```

```python
grouped = engage_group.drop('user_id',axis=1)
grouped.head()
```

```python
# Reset the index.
grouped.reset_index(inplace=True)

# Sum the total number of active weeks for each user_id
grouped_users = grouped.groupby('user_id').sum()

# Create adopted_user columns, if users have more than 3 visits assign
# else assign adopted_user =  
grouped_users['adopted_user']=grouped_users['visited'].apply(lambda x: 1 if x>3 else 0)
```

```python
# Reset index.
grouped_users.reset_index(inplace=True) 
```

```python
grouped_users.head()
```

```python
grouped_users.adopted_user.value_counts()
```

### Missing values

Let's explore some missing values in the dataframe. 

```python
# Null value from last_session_creation_time.
merged[merged.last_session_creation_time.isnull()].head()
```

```python
# Null value from created_session
merged[merged.invited_by_user_id.isnull()].head()
```

There are different number of values on last_session_creation_time and invited_by_user_id columns. For __'last_session_creation_time'__, there are about more than 300 users that have an account but never created a session. And __'invited_by_user_id'__ columns, the missing values can assume that the users were not invited by someone else.

So, I will create a new column __'created_session'__ which tells if the user ever create a session get value of 1, otherwise receives value of 0. And for null values in 'last_session_creation_time', if the users do not have last session, it means the last session he created has the same time as the creation_time. So it is reasonable if we fill null value with creation_time.

```python
# 'create_session' if last_session_creation_time is not null.
merged['create_session']=merged['last_session_creation_time'].notnull().astype(int)
```

For invited_by_user_id, because if user is not invited by nobody the value will be null. We will fill the null value with 0's. Else, if there is an id, we replace the values with 1.

```python
# Fill the null values for the last_session_creation_time
merged['last_session_creation_time'].fillna(merged.creation_time, inplace=True)

# Fill the null value for invited_by_user_id
merged.invited_by_user_id.fillna(0,inplace=True)
merged['is_invited'] = merged.invited_by_user_id.apply(lambda x:1 if x!=0 else 0)

# Fill the null values for visited
merged.visited.fillna(0,inplace=True)

# Fill the null values for adopted_user
merged.adopted_user.fillna(0,inplace=True)
```

### Create new dataset
Now, we create a new dataframe to prepare the data for our model. Because there are features that are not really meaningful to the model, so I will remove those features like 'name','email,', 'provider',etc and use the encoded features instead.

```python
# Remove unnessary features.
new_df=merged.drop(['object_id', 'creation_time', 'name', 'email', 'creation_source', 'last_session_creation_time', 'invited_by_user_id','create_session','email_provider','org_id','visited',],axis=1)
```

```python
list(new_df.columns)
```

### Model implementation




Now we have a dataset, so we will perform some modelling and check feature importance. Because the model is imbalanced, so we will use SMOTE to upscale the training data set.



```python
# Create data into X(variables) and y(target)
X=new_df.drop('adopted_user',axis=1)
y=new_df['adopted_user']

X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=0,test_size=0.25)
```

```python
#Check the balance of target variable
y_train.value_counts()
```

Because the response variable is very imbalanced so we will use SMOTE to upscale the number of variables.

```python
# Create SMOTE model
sm = SMOTE(random_state=0)

# Resampling dataset
X_res, y_res = sm.fit_resample(X_train.values, y_train)

print('Resample dataset shape', Counter(y_res))
```

With the resampled dataset, we will fit the model using RandomForestClassifier. 

```python
# RandomForestClassifier
rf = RandomForestClassifier(n_estimators=100)

# Fit the model
rf.fit(X_train, y_train)
```

```python
predicted = rf.predict(X_test)
```

```python
# Classification report
print(classification_report(y_test, predicted))
```

```python
# Create a dataframe of feature importance
feature_importance = pd.DataFrame(rf.feature_importances_).transpose()
```

```python
feature_importance
```

```python
# Combine the feature importance with column name
feature_importance.columns = list(X_train.columns)
feature_importance.index=['Feature Importance']
feature_importance=feature_importance.transpose().sort_values(by='Feature Importance',ascending=False)
```
