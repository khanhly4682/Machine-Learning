```python
# Splitting data into training and test set
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

X = df[['Height','Weight']].values
#y = [df['Gender']=='Male']
X_train, X_test, y_train, y_test = train_test_split(X, (df.Gender == 'Male').values, random_state=5)

# Create LogisticRegression model.
reg = LogisticRegression()

# Fit data.
reg.fit(X_train, y_train)

# Print the accuracy from the testing data.
predict = reg.predict(X_test)
accuracy = accuracy_score(predict,y_test)
print(accuracy)
```

```python
# Cross validation.
from sklearn.model_selection import KFold
from sklearn.metrics import accuracy_score

#Define  K-fold cross validation.
#def cv_score1(reg, x, y, score_function=accuracy_score):
#    result = 0
#    nfold = 5
#    for train_indx, test_indx in KFold(nfold).split(x):
#        reg.fit(x[train_indx], y[train_indx])
#        result += score_function(reg.predict(x[test_indx]), y[test_indx])
#    return result / nfold # Get average.

def cv_score(clf, x, y, score_func=accuracy_score):
    result = 0
    nfold = 5
    for train, test in KFold(nfold).split(x): # split data into train/test groups, 5 times
        clf.fit(x[train], y[train]) # fit
        result += score_func(clf.predict(x[test]), y[test]) # evaluate score function on held-out data
    return result / nfold # average

# Using cv_score function for basic logistic regression without regularization.
reg = LogisticRegression()
score = cv_score(reg, X_train, y_train)
print(score)
```

Your goal is to find the best model parameters based *only* on the training set, without showing the model test set at all (which is why the test set is also called a *hold-out* set).

```python
#the grid of parameters to search over
# C is the regularization parameter.
# Smaller C specify stronger regularization.
Cs = [0.001, 0.1, 1, 10, 100]

for c in Cs:
    reg = LogisticRegression(C=c)
    score = cv_score(reg, X_train, y_train)
    print('C: ' + str(c) + '   score: ' + str(score))
```

<b>GridSearchCV():</b>
* Grid search builds a model for every combination of hyperparameters specified and evaluates each model
* Exhaustive search over specified parameter values for an estimator.
* GridSearchCV lets you combine an estimator with a grid search preamble to tune hyper-parameters. 
* The method picks the optimal parameter from the grid search and uses it with the estimator selected by the user.

```python
from sklearn.model_selection import GridSearchCV

grid = GridSearchCV(estimator=LogisticRegression(), cv=5, param_grid = {'C': [0.001, 0.1, 1, 10, 100]})
grid.fit(X_train, y_train)

accuracy = accuracy_score(grid.predict(X_test), y_test)

print(accuracy)
```
