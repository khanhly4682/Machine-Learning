### Has the client subscribed a term deposit? ('yes', 'no')

```python
data['y'].value_counts()
```

```python
# Calculate percentage of term deposit subscription
subscribed = round(data[data['y']=='yes']['y'].count() / data['y'].count(), 2) * 100
un_subscribed = round(data[data['y']=='no']['y'].count() / data['y'].count(), 2) * 100
print("Subscribed:", subscribed, "%")
print("Un_subcribed:", un_subscribed, "%")
```

```python
# Plot Term Deposit Subscription Percentage
fig, ax = plt.subplots()

labels = ['Subscribed', 'Unsubscribed']
percentages =[subscribed,un_subscribed]
explode = (0.1, 0)

ax.pie(percentages, explode=explode, labels = labels,autopct='%1.0f%%')
ax.set_title('Term Deposit Subscription')
plt.show()
```

There is 11% of customers subcribed to term deposit and 89% did not subcribe. It shows that this is unbalanced dataset which we will have to pay attention to when we implement the predictive models.

### 2.4 Missing values

```python
missing_vals = df.isnull().sum().sort_values(ascending=False)
missing_vals
```

The data has no missing values (e.g. null/na values). However, there are features contain **unknown** values that we will deal with those values. There are different ways to deal with missing values, for example discarding the data if the data portion is large (like 70% of the corpus), or we can do imputation

Categorical Features that have unknown values: **['job', 'marital','education','default','housing','loan']**<br>


#### Unknown value - 'job'
There are 330 unknown value in 'job' columns; and 785 people whose ages over 60 are retired. Assumed that people who aged more than 60 will be retired, which is reasonable in the real life. We will replace job as 'retired' for the group of unknown jobs.

```python
# Count the number of 'unknown' job/
df[df['job']=='unknown']['job'].count()
```

```python
# Count the numbers of job that people who are over 60 years old do
df['job'][df['age']>=60].value_counts()
```

```python
# Replace unknown job where age >=60
df.loc[(df['age']>=60) & (df['job']=='unknown'), 'job'] = 'retired'
```

```python
# After replacing 'retired' for unknown group which aged >=60
df['job'][df['age']>=60].value_counts()
```

There will be a relationship between job and education. For example, 'admin', 'management', and 'technician' normally have univerisity degree. Most of technicians have professional development.

We can infer the job from education and education based on job title. For example:

- Admins have univeristy education and do not have loan or default.
- Blue collar have 4 year, 6 year, 9 year education
- Services have high school education
- Housemaid have 4 year education


```python
# Replace unknown values as admin
df.loc[(df['job']=='unknown') 
       & (df['education']=='university.degree') 
       & (df['loan']=='no')
       , 'job'] = 'admin'
```

```python
# Replace unknown values as technician
df.loc[(df['job']=='unknown') 
       & (df['education']=='professional.course') 
       & (df['loan']=='no')
       , 'job'] = 'technician'
```

```python
# Replace unknown values as blue_colar
df.loc[(df['job']=='unknown') & (df['education']=='basic.4y'), 'job'] = 'blue-collar'
df.loc[(df['job']=='unknown') & (df['education']=='basic.6y'), 'job'] = 'blue-collar'
df.loc[(df['job']=='unknown') & (df['education']=='basic.9y'), 'job'] = 'blue-collar'
```

```python
# Replace unknown values for education f.loc[(df['job']=='unknown') & (df['education']=='basic.4y'), 'job'] = 'blue-collar'
df.loc[(df['education']=='unknown') & (df['job']=='admin.'), 'education'] = 'university.degree'
df.loc[(df['education']=='unknown') & (df['job']=='technician'), 'education'] = 'professional.course'
df.loc[(df['education']=='unknown') & (df['job']=='housemaid'), 'education'] = 'basic.4y'
df.loc[(df['education']=='unknown') & (df['job']=='management'), 'education'] = 'university.degree'
df.loc[(df['education']=='unknown') & (df['job']=='entrepreneur'), 'education'] = 'university.degree'
```
